```python
from scipy.integrate import quad
```

```python
def u(x):
    #runge function
    return 1/(1+50*(0.5-x)**2)
```

```python
x=linspace(0,1,1025)
```

```python
plot(x,u(x))
```

```python
#basis functions
n=10

def V(x,i):
    return x**i
```

```python
_=[plot(x,V(x,i)) for i in range(n)]
```

```python
M = zeros((n,n))
for i in range(n):
    for j in range(n):
        M[i,j] = quad(lambda x: V(x,i)*V(x,j), 0, 1)[0]
```

```python
F = array( [quad( lambda x: V(x,i)*u(x), 0, 1 )[0] for i in range(n)])
```

```python
p=linalg.solve(M,F)
```

```python
Vx=array( [V(x,i) for i in range(n)]).T
```

```python
plot(x,u(x))
plot(x,Vx.dot(p))
```

```python
def legendre(x,i):
    if i==0:
        return x*0 + 1
    if i==1:
        return 2*(x-0.5)
    else:
        n=i-1
        return ((2*n+1)*x*legendre(x,n) - n*legendre(x,n-1))/(n+1)
```

```python
x=linspace(-1,1,1025)
[plot(x,legendre(x,i)) for i in range(n)]
```
